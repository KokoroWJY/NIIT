### 作业内容：宝可梦数据处理

```python
import pandas as pd
import numpy as np
```

```python
# 读取宝可梦数据，目录：./data/Pokemon.csv
data = pd.read_csv('./data/Pokemon.csv')
# 查看前五条数据
data.head(5)
data[0: 5]
```

```python
# 显示DataFrame数据的每列的数据类型和非空数据
data.info()
```

```python
# 显示DataFrame的描述性统计数据
data.describe()
```

```python
# 计算Type 1 字段共有几种取值，分别有多少数量
data['Type 1'].value_counts()
```

```python
# 查看'Type 1'的值为 'Grass'的数据
data[data['Type 1'] == 'Grass']
```

```python
# 查看 total > 700 且 Attack > 100 的数据
data[(data['Total'] > 700) & (data['Attack'] > 10)]
```

```python
# 查看'Name'字段以'Ka'开头的数据
data[data['Name'].str.startswith("Ka")]
```

```python
# 新增一个列，命名为'new_col'，值为'HP'数值的1/100
data['new_col'] = data['HP'] / 100
data
```

```python
# 删除刚才创建的列'new_col'
data = data.drop(['new_col'], axis=1)
# data.drop(['new_col'], axis=1, inplace=True)
data.head()
```

```python
# 添加一个新列new_col_2，布尔类型，当以下两项均满足时为True，否则为False
# 条件1：Name 以 'B'开头。  条件2：Type 2 为空。
data['new_col_2'] = data['Name'].str.startswith('B') & (data['Type 2'].isnull())
print(data['new_col_2'].value_counts())
data[data['new_col_2'] == True]
data[data['new_col_2'] == True].shape
```

```python
# 显示'new_col_2'中为True的数据
data['Type 2'].isnull().sum()
```

```python
# 查看Type 2 为空的记录数
# isna(), isnull(), dropna(), fillna()
data['Type 2'].fillna('Unknown', inplace=True)
```

```python
# 将Type 2的空字段，填充为'Unknown'
data['Type 2'].isnull().sum()
```

```python
# 再来查看Type 2 为空的记录数

data['Type 2'].value_counts()
```
