## 1. 折线图  plt.plot

### 1.1 先来一个最基础的

```python
# 画一个x平方函数
# 数据准备: x:[-5, 5] 取21个点, y = x * x
x = np.linspace(-5, 5, 21) #采样点越多越接近理想曲线
y = x ** 2
plt.plot(x, y)
```

```python
# 练习：画sin函数 pi：np.pi, x轴: 0-2pi
x = np.linspace(0, 2 * np.pi, 101)
y = np.sin(x)
plt.plot(x, y)
```

### 1.2 学习一个完整的绘图过程

```python
# (1) 生成一个画布
plt.figure(figsize=(10, 5), # 画布大小 (长, 高)
          dpi=300)

# (2) 生成数据. x: 0-4, y: [12, 7, 4, 8,2]
x = np.arange(5)
y = np.array([12, 7, 5, 3, 1])

# (3) 选择绘图类型, 折线图 plt.plot() color: 线的颜色, marker:标记类型, linestyle: 线的类型, markersize: 标记大小
plt.plot(x, y, color='r', marker='+', linestyle='--', markersize=12)

# (4) 添加标题
plt.title('折线图实例1')

# (5) 设置x轴y轴的名称
plt.xlabel('人数(万)')
plt.ylabel("销量(万元)")

# (6) 其他美化
# 增加x轴标注
plt.xticks(ticks=np.arange(5), # 标记的位置
           labels=['1月','2月','3月','4月','5月']) # 显示的内容
# 增加背景网格
plt.grid()

# 最后,展现所有的内容
plt.show()
```

### 1.3 尝试一张图上多条折线

```python
# (1) 生成一个画布
plt.figure(figsize=(10, 5), # 画布大小 (长, 高)
          dpi=None)
# (2) 生成数据. x: 0-4, y: [12, 7, 4, 8,2]
x = np.arange(5)
y = np.array([12, 7, 4, 8, 2])
y2 = np.array([8, 4, 15, 2, 5])

# (3) 选择绘图类型, 折线图 plt.plot() 通过label设置每一条线的名称
plt.plot(x, y, color='c', marker='+', linestyle='--', markersize=12, label='门店1')
plt.plot(x, y2, color='b', marker='o', linestyle='-', markersize=6, label='门店2')

# (4) 添加标题
plt.title('折线图实例1')

# (5) 设置x轴y轴的名称
plt.xlabel('人数(万)')
plt.ylabel("销量(万元)")

# (6) 其他美化
plt.xticks(ticks=np.arange(5), labels=['1月','2月','3月','4月','5月'])
plt.grid()

# 增加图例 两条以上, 需要
plt.legend(loc=0) # 默认自动寻找最好的位置

# (7) 最后,展现所有的内容
plt.show()
```

### 1.4 课堂练习：画出sin和cos的图形，x：0-2pi

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5), dpi=100)
plt.plot(x, y1, color='r', label='sin')
plt.plot(x, y2, color='b', label='cos')
plt.title("sin和cos函数图")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(ticks=[0, 0.5*np.pi, np.pi, 1.5*np.pi, 2 * np.pi], labels=['0', '0.5pi', 'pi', '1.5pi', '2pi'])
plt.grid()
plt.legend(loc=3)
plt.show()
```

## 2. 散点图  plt.scatter

```python
x = np.random.random(30) # 30个0-1之间的随机数
y1 = np.random.random(30)
y2 = np.random.random(30)+0.5

plt.figure(figsize=(8, 4))
# 画散点图 alpha: 透明度, linewidths: 散点的大小
plt.scatter(x, y1, color='r', alpha=0.5, linewidths=3, marker='o', label='y1')
plt.scatter(x, y2, color='g', alpha=0.5, linewidths=3, marker='v', label='y2')

plt.legend()
plt.show()
```

## 3. 直方图 plt.hist
- 直方图：在统计数据时，按照频数分布表，在平面直角坐标系中，横轴标出每个组的端点，纵轴表示频数，每个矩形的高代表对应的频数，称这样的统计图为频数分布直方图。
- ①能够显示各组频数分布的情况； ②易于显示各组之间频数的差别。

### 3.1 了解一下什么是直方图

```python
# 生成一个数组：[1,1,1, 2, 3, 4,4, 5,5,5,5]
x = np.array([1,1,1, 2, 3, 4,4, 5,5,5,5])


# 分组统计(分箱) 等宽分箱: 宽度=1, [1,2), [2,3), [3,4), [4,5]
# 直方图: 每个箱子里盛放的数字的数量: [3], [1], [1], [6]
plt.hist(x, bins=4) # bin箱子个数
```

### 3.2 试试随机产生的数据的分布

```python
# 随机生成0-100之间的数字，1000个，看看分布
x = np.random.randint(0, 100, 1000)
plt.hist(x, bins=10) # 默认: 10-个分箱
```

### 3.3 看看正态分布
- 正态分布（高斯分布） [百度百科](https://baike.baidu.com/item/%E6%AD%A3%E6%80%81%E5%88%86%E5%B8%83/829892?fr=aladdin)

## 4. 在一个图表中画多个子图 plt.subplot

```python
# 上下两个图，上：log函数，下：exp函数

# 图1：log函数
x1 = np.linspace(0.01, 5, 20)
y1 = np.log(x1)

# 图2： exp函数
x2 = np.linspace(0, 2, 20)
y2 = np.exp(x2)


# （1）生成画布
plt.figure(figsize=(8, 8))

# （2）画第一个子图 nrow 行, ncols 列, index当前画第几个
plt.subplot(2, 1, 1) # 行, 列, 当前画第1个子图(index从1开始)
plt.plot(x1, y1, 'bo-') # 简写法: 颜色标记线迹
plt.title('图1 log函数')

# （3）画第二个子图
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'rv-')
plt.title('图2 exp函数')

# 设置间隔 subplots_adjust, wspace横向间隔 hspace纵向间距
plt.subplots_adjust(hspace=0.5)
```

#### 附1：关于设置中文支持
matplotlib默认情况下不支持中文显示，如果需要显示中文，则我们需要做一些额外的设置操作。设置可以分为：

我们可以通过执行：  
`mpl.rcParams["font.family"] = "中文字体名称"`  
`mpl.rcParams["axes.unicode_minus"]=False`  
进行设置。常用的设置如下:
* font.family 字体的名称
 + SimHei 中文黑体
 + FangSong 中文仿宋
 + YouYuan 中文幼圆
 + STSong 华文宋体
 + Kaiti 中文楷体
 + LiSu 中文隶书
* font.style 字体的风格
 + normal 常规（默认）
 + italic 斜体
 + oblique 倾斜
* font.size 字体的大小（默认10）
* axes.unicode_minus 是否使用Unicode的减号（负号）【在支持中文显示状态下，需要设置为False】

#### 附2：关于画子图的另一种写法

```python
# 画多个子图方法二：
fig, ax = plt.subplots(2)
ax[0].plot(x1, y1, 'ko-')
ax[0].set(title='A tale of 2 subplots', ylabel='Damped oscillation')

ax[1].plot(x2, y2, 'r.-')
ax[1].set(title='A tale of 2 subplots',xlabel='time (s)', ylabel='Undamped')

plt.show()
```
