## 一、关键词提取

### 1、Jieba分词中的关键词提取功能

```python
import jieba
import jieba.analyse as aly
from gensim import corpora, models
```

```python
#读取文本信息
path = '481681.txt'
def get_text(text):
    with open(path,'r',encoding='utf-8') as f:
        text = ''
        text = f.read()
        return text
text = get_text(path)
print('初始文本:',text)
```

#### (1)使用jieba中TF-IDF的extract_tags()方法

```python
#加载停用词典
aly.set_stop_words('stop_words.txt')

# 第一个参数：待提取关键词的文本
# 第二个参数：TF/IDF 权重最大返回关键词的数量，重要性从高到低排序，默认值为 20
# 第三个参数：是否同时返回每个关键词的权重，默认值为 False
# 第四个参数：词性过滤，为空表示不过滤（默认），若提供则仅返回符合词性要求的关键词
keywords = aly.extract_tags(text, topK=10, withWeight=True, allowPOS=())
print(keywords)

for item in keywords:
    # 分别为关键词和相应的权重
    print(item[0], item[1])
    
    
#jieba.analyse.TFIDF(idf_path=None) #新建 TFIDF 实例，idf_path 为 IDF 频率文件
```

#### 使用tfidf()方法

```python
# 第一个参数：待提取关键词的文本
# 第二个参数：TF/IDF 权重最大返回关键词的数量，重要性从高到低排序，默认值为 20
# 第三个参数：是否同时返回每个关键词的权重，默认值为 False
# 第四个参数：词性过滤，为空表示不过滤（默认），若提供则仅返回符合词性要求的关键词
keywords2 = aly.tfidf(text, topK=10, withWeight=True, allowPOS=())
print(keywords2)

for item in keywords2:
    # 分别为关键词和相应的权重
    print(item[0], item[1])
```

#### (2)使用textrank()方法

**不需要数据集**

```python
# 第一个参数：待提取关键词的文本
# 第二个参数：返回关键词的数量，重要性从高到低排序
# 第三个参数：是否同时返回每个关键词的权重
# 第四个参数：词性过滤，为空表示过滤所有，与TF—IDF不一样！
#allowPOS('ns', 'n', 'vn', 'v') 地名、名词、动名词、动词
keywords = jieba.analyse.textrank(text, topK=10, withWeight=True)
for item in keywords:
    # 分别为关键词和相应的权重
    print(item[0], item[1])
```

### 2.LSI和LDA用法示例

```python
import re
 
# 生成LDA模型
def LDA_model(words_list):
    # 构造词典
    # 单词出现次数以及相关的统计信息
    dictionary = corpora.Dictionary(words_list)
    print('查看每个单词的id:')
    print(dictionary.token2id)  # 打印查看每个单词的id
 
    # 在每个文档向量中都包含一系列元组,元组的形式是（单词 ID，词频）
    corpus = [dictionary.doc2bow(words) for words in words_list]
    print('输出每个文档的向量:')
    print(corpus)  # 输出每个文档的向量
 
    # LDA主题模型
    # num_topics -- 必须，要生成的主题个数。
    # id2word    -- 必须，LdaModel类要求我们之前的dictionary把id都映射成为字符串。
    # passes     -- 可选，模型遍历语料库的次数。
    lda_model = models.ldamodel.LdaModel(corpus=corpus, num_topics=20, id2word=dictionary, passes=10)
 
    return lda_model

# 生成LSI模型
def LSI_model(words_list):
    # 构造词典
    # 单词出现次数以及相关的统计信息
    dictionary = corpora.Dictionary(words_list)
     # 在每个文档向量中都包含一系列元组,元组的形式是（单词 ID，词频）
    corpus = [dictionary.doc2bow(words) for words in words_list]
  
    # LSI主题模型
    # num_topics -- 必须，要生成的主题个数。
    # id2word    -- 必须，LdaModel类要求我们之前的dictionary把id都映射成为字符串。
    # passes     -- 可选，模型遍历语料库的次数。
    lsi_model = models.lsimodel.LsiModel(corpus=corpus, num_topics=20, id2word=dictionary)
 
    return lsi_model

# 分词处理
out_text = ''
text = re.sub('\u3000','',text)
text = text.replace('\n','')
sent_text = jieba.cut(text.strip())
print('\n','分词结果：',sent_text)
#print('\n','分词结果：','\n'+'/ '.join(sent_text)+'\n')


# 去停用词
stopwords = open('stop_words.txt',encoding="utf-8").read().split('\n')
for word in sent_text:
    if word not in stopwords:
        if word != '\t':
            out_text += word
            out_text += ' '

# # 去掉只出现一次的单词
print('\n','去停用词后结果：',out_text) 


# 字符串转为列表
out_list = []
list = ''
for i in out_text:
    if i != ' ':
        list = list + i
    else:
        out_list.append(list)
        list = ''

print('\n','转换为列表：',out_list,'\n')
        
# 获取训练后的LDA模型
# doc2bow expects an array of unicode tokens on input, not a single string，加个[]，使其变成列表
lda_model = LDA_model([out_list])
lsi_model = LSI_model([out_list])

# 输出
topicwords_lda = lda_model.show_topic(0, 10)
print('\n','LDA主题模型关键词列表：',topicwords_lda)

topicwords_lsi = lsi_model.show_topic(0, 10)
print('\n','LSI主题模型关键词列表：',topicwords_lsi)
```
